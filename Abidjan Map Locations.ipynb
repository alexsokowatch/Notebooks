```python
import folium
import pandas as pd
import numpy as np
```

```python
abj_dukas = pd.read_csv("C:/Users/alexd/Downloads/Duka-cocody- POS.csv", header=0)
```

```python
abj_dukas.head()
```

```python
locations = abj_dukas[['Geo Latitude', 'Geo Longitude']]
locationlist = locations.values.tolist()
```

```python
map = folium.Map(location=[abj_dukas['Geo Latitude'].mean(), abj_dukas['Geo Longitude'].mean()], zoom_start=12)
for point in range(0, len(abj_dukas)):
    folium.Marker(locationlist[point], popup=abj_dukas['Outlet Name'][point]).add_to(map)
map
```

```python
map.save("C:/Users/alexd/Downloads/cocody_map.html")
```
